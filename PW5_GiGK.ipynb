# Współrzędne jednorodne w $\mathbb{R}^3$


W zbiorze $A=\mathbb{R}^4-\{0,0,0,0\}$ określmy relacje ~ :

$(x_1,y_1,z_1,w_1)$~$(x_2,y_2,z_2,w_2)$ $\iff$ istnieje $r>0$ takie, że $(x_1,y_1,z_1,w_1)=r(x_2,y_2,z_2,w_2)$

Jest to relacja równoważności w $\mathbb{R}^4$. Zbiór klas abstrakcji tej relacji nazywamy płaszczyzną rzutową 

## Przykłady

Odwzorowanie liniowe:

$\begin{bmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{bmatrix}\to
\begin{bmatrix}
a & b & c & 0\\
d & e & f & 0\\
g & h & i & 0\\
0 & 0 & 0 & 1
\end{bmatrix}$

Translacja o wektor $(p,q, r)$:

$\begin{bmatrix}
1 & 0 & 0 & p \\
0 & 1 & 0 & q\\
0 & 0 & 1 & r\\
0 & 0 & 0 & 1
\end{bmatrix}$

# Prosta w $\mathbb{R}^2$ a płaszczyzna w $\mathbb{R}^3$

Prosta w $\mathbb{R}^2$ jest analogią płaszczyzny w $\mathbb{R}^3$

Prosta w $\mathbb{R}^2$ we współrzędnych jednorodnych dana jest równaniem $Ax+By+Cw=0$.

Równoważnie: $(A,B,C)\circ(x,y,w)=0$

Jeśli $(x_1,y_1,w_1),(x_2,y_2,w_2)\in l$, to $\alpha_1(x_1,y_1,w_1)+\alpha_2(x_2,y_2,w_2)\in l$

$l: (x,y,w)=\alpha_1(x_1,y_1,w_1)+\alpha_2(x_2,y_2,w_2)$ jest postacią "parametryczną" prostej przechodzącej przez dwa punkty (dwa parametry)

Jeśli chcielibyśmy mieć zależność od jednego parametru (klasyczną parametryzację), to możemy to zrobić biorąc $w_1=w_2=1$, $\alpha_1+\alpha_2=1$

Mając dane dwa punkty  $(x_1,y_1,w_1),(x_2,y_2,w_2)$ można też obliczyć (A,B,C) jako iloczyn wektorowy
$$(A,B,C)=
\begin{vmatrix}
e_1 & e_2 & e_3 \\
x_1 & y_1 & w_1\\
x_2 & y_2 & w_2
\end{vmatrix}$$
gdzie $(e_1, e_2, e_3)$ - baza standardowa

## Płaszczyzna w $\mathbb{R}^3$

Równanie płaszczyzny w $\mathbb{R}^3$ ma taką samą strukturę jak równanie prostej w $\mathbb{R}^2$

PODOBNIE JAK W PRZYPADKU PROSTEJ

Płaszczyzna w $\mathbb{R}^3$ we współrzędnych jednorodnych dana jest równaniem $Ax+By+Cz+Dw=0$.

Równoważnie: $(A,B,C,D)\circ(x,y,z,w)=0$

$(x,y,z,w)=\alpha_1(x_1,y_1,z_1,w_1)+\alpha_2(x_2,y_2,z_2,w_2)+\alpha_3(x_3,y_3,z_3,w_3)$ jest postacią "parametryczną" płaszczyzny przechodzącej przez trzy dane punkty

Wówczas
$$(A,B,C,D)=
\begin{vmatrix}
e_1 & e_2 & e_3 & e_4\\
x_1 & y_1 &  z_1& w_1\\
x_2 & y_2 & z_2& w_2 \\
x_3 & y_3 & z_3& w_2 
\end{vmatrix}$$
gdzie $(e_1, e_2, e_3,e_4)$ - baza standardowa

RÓŻNICE:
- 3 płaszczyzny  $A_ix+B_iy+C_iz+D_iw=0$, $i=1,2,3$
 przecinają się w punkcie
$$(x,y,z,w)=
\begin{vmatrix}
e_1 & e_2 & e_3 & e_4 \\
A_1 & B_1 & C_1 & D_1\\
A_2 & B_2 & C_2 & D_2\\
A_3 & B_3 & C_3 & D_3\\
\end{vmatrix}$$
- 2 płaszczyzny nie przecinają się w punkcie, ale wzdłuż prostej

$$(A_1,B_1,C_1,D_1)\circ(x,y,z,w)=0$$ i $$(A_2,B_2,C_2,D_2)\circ(x,y,z,w)=0$$




Przykładowo układ ten możemy rozwiązać następująco:




(prosta - kombinacja dwóch punktów)
 $(x,y,z,w)=\alpha(x_1,y_1,z_1,w_1)+\beta(x_2,y_2,z_2,w_2)$
 
 
Skąd taki wzór?

Dla $w=0$:


$(A_1,B_1,C_1)\times (A_2,B_2,C_2)$

Dla $z=0$:


$(A_1,B_1,D_1)\times (A_2,B_2,D_2)$

Ale oczywiście nie jest to jedyny możliwy sposób


PROBLEM: jeśli płaszczyzny się przecinają, to mogę dostać dwa wektory zależne. Albo któryś z tych wektorów jest zerowy.  Wtedy np. zamiast $z=0$ można wybrać np. $x=0$. Dobrze jednak nie zmieniać pierwszego wektora. Jeśli $(A_1,B_1,C_1)=0$, to mamy płaszczyzny równoległe.


PRZYPOMNIENIE: żeby wrócić do klasycznej geometrii musimy podzielić przez $w$.

Punkt w nieskończoności można traktować jako wspólny element wszystkich prostych o ustalonym kierunku (nie z wektorem, ale z kierunkiem (nie ma zwrotu))

PROBLEM :  $(x,y,z,w)=\alpha(x_1,y_1,z_1,w_1)+\beta(x_2,y_2,z_2,w_2)$

Żeby w taki sposób opisać prostą muszę podać 8 liczb, za dużo, nie wyznaczają prostej jednoznacznie
