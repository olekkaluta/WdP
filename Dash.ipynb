```python
from dash import dcc
from dash import Dash, html, dcc
import plotly.graph_objects as go
from dash.dependencies import Input, Output


app = Dash(__name__)

app.layout = html.Div([
    html.Label('Wybierz a:'),
    html.Br(),
    dcc.Slider(0, 10, 1, value=2, id='my-slider'),
    html.Br(),
    html.Label('Wybierz b:'),
    html.Br(),
    dcc.Slider(0, 10, 1, value=3, id='my-slider2'),
    html.Br(),
    html.Label('Wybierz p:'),
    html.Br(),
    dcc.Slider(-10, 10, 1, value=0, id='my-slider3'),
    html.Br(),
    html.Label('Wybierz q:'),
    html.Br(),
    dcc.Slider(-10, 10, 1, value=0, id='my-slider4'),
    html.Br(),
    html.Label('Wybierz n:'),
    html.Br(),
    dcc.Slider(0, 10, 1, value=1, id='my-slider5'),
    html.Div(
    dcc.Graph(id='chart'))

], style={'padding': '30px'})

@app.callback(
    Output('chart', 'figure'),
    Input('my-slider', 'value'),
    Input('my-slider2', 'value'),
    Input('my-slider3', 'value'),
    Input('my-slider4', 'value'),
    Input('my-slider5', 'value'))

def update_output(a, b, p, q, n):
    fig = go.Figure()
    fig.update_xaxes(range=[-10, 10])
    fig.update_yaxes(range=[-10, 10])
    fig.add_shape(type="circle",
    xref="x", yref="y",
    x0=n*(-a+p), y0=n*(-b+q), x1=n*(a+p), y1=n*(b+q),
    line_color="LightSeaGreen",
)
    return fig

if __name__ == '__main__':
    app.run_server(debug=False)
```
