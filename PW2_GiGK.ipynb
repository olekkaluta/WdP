```python
#Zaimoportować bliblioteki numpy i matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
    #Zdefiniować dwa wektory (użyć numpy)
#https://numpy.org/doc/stable/user/absolute_beginners.html
x = np.vstack([1,1])
y = np.vstack([1,2])
```

```python
plt.axes().arrow(0,0,x[0][0],x[1][0],head_width=0.1, head_length=0.1)
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.grid(alpha=1)

plt.title(r"Wektor x=[%d," %x[0]+ "%d]" %x[1]  )
plt.show()
```

```python
#  Policzyć kąt pomiędzy wektorami
cosAngle= np.dot(x.T,y)/np.sqrt(np.dot(np.dot(x.T,x),np.dot(y.T,y)))
print(cosAngle)

angle=np.arccos(cosAngle)*(180/np.pi)

print(angle)
print("Kąt między wektorami ",x.T, " ",y.T," wynosi ", angle[0][0])
```

```python
##W układzie współrzędnych narysować dwa wektory. W tytule wykresu podać współrzędne wektorów.
#Na wykresie umieścić wartość kąta pomiędzy wektorami.


plt.arrow(0,0,x[0][0],x[1][0],head_width=0.1, head_length=0.1)
plt.arrow(0,0,y[0][0],y[1][0],head_width=0.1, head_length=0.1)

plt.xlim(-3,3)
plt.ylim(-3,3)

plt.annotate(str(round(angle[0][0]))+r"$^\circ$",xy=(x[0][0]/2,y[1][0]/2));


plt.grid(alpha=1)

plt.title(r"Wektor [%d," %x[0]+ "%d]" %x[1] +" oraz wektor [%d," %y[0]+ "%d]" %y[1] )
plt.show()
```

```python
# Obrót o kąt pi/4
fi=np.pi/4

obr=np.vstack([[np.cos(fi),-np.sin(fi)],[np.sin(fi),np.cos(fi)]])
print(obr)

rex=np.matmul(obr,x)
print(rex)
plt.arrow(0,0,x[0][0],x[1][0],head_width=0.1, head_length=0.1)
plt.arrow(0,0,rex[0][0],rex[1][0],head_width=0.1, head_length=0.1,color='blue')

plt.xlim(-3,3)
plt.ylim(-3,3)

plt.grid(alpha=1)
```

```python
# - skalowanie (deformacja) o współczynnikach (2,3) względem zera 
ms=np.vstack([[2,0],[0,3]])
print(ms)

sk=np.matmul(ms,x)
print(sk)
plt.arrow(0,0,x[0][0],x[1][0],head_width=0.1, head_length=0.1)
plt.arrow(0,0,sk[0][0],sk[1][0],head_width=0.1, head_length=0.1,color='blue')

plt.xlim(0,4)
plt.ylim(0,4)

plt.grid(alpha=1)
```

```python

# *********************** LINIOWA NIEZALEŻNOŚĆ ***********************


#1 Czy dwa wektory z powyższych zadań (x i y) są liniowo niezależne 
import numpy as np
import matplotlib.pyplot as plt

x = np.vstack([1,1])
y = np.vstack([1,2])

wyz= np.dot(x[0][0],y[1][0])-np.dot(x[1][0],y[0][0])

print(wyz)
if(wyz==0):
    print("Wektory x, y są zależnny")
else:
    print("Wektory x, y są niezależne")
```

```python

# *********************** BAZA ORTONORMALNA ***********************

#2 Unormować jeden z wektorów z poprzedniego zadania (x) i zapisać go jako b1

#3 Utworzyć wektor b2 prostopadły do wektora b1

#4 Jakie są współrzędne wektora y w bazie (b1,b2) ?



dlx=np.sqrt(np.dot(x.T,x))
print(dlx)
b1=x/dlx
print(b1)

fi=np.pi/2

obr=np.vstack([[np.cos(fi),-np.sin(fi)],[np.sin(fi),np.cos(fi)]])
print(obr)

b2=np.matmul(obr,b1)
print(b2)
```
