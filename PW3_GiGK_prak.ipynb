```python
import numpy as np
import matplotlib.pyplot as plt
import math

plt.figure()
plt.axis([0,100,100,0])

r=40
alpha1=0
alpha2=2*math.pi
dalpha=math.pi/180
x=[]
y=[]
xc=50
yc=50

for alpha in np.arange(alpha1,alpha2,dalpha):
    x.append(xc+r*math.cos(alpha))
    y.append(yc+r*math.sin(alpha))
plt.scatter(x,y,s=5,color='k')
plt.show()
```

## Pierwszy przykład w $\mathbb{R}^3$

```python


from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x =[1,2,4,6]
y =[5,6,2,7]
z =[2,3,2,8]

ax.scatter(x, y, z, c='r', marker='o')
ax.plot
#scatterplot of points

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()

```

### ĆWICZENIE 2 
Narysować  wykres funkcji $f(x,y)=x^2-y^2$ 

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

def f(x, y):
    return x*x-y*y

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
x, y = np.meshgrid(x, y)
z=f(x, y)
ax.scatter3D(x, y, z);
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.plot

plt.show()
```
